```python
# Importer les différentes bibliothèques de traitement des données
import pandas as pd, numpy as np
```

```python
# Lire le fichier CSV prétraité
data  = pd.read_csv(r"./final_dataset.csv", sep = ";")
```

```python
# Afficher le résultat
data
```

```python
# Creer une variable intervales d'ages qui classesnt les personnes selon la plage d'age ou elles appartiennet 
intervales_ages = pd.cut(data.age, bins=[0,18,25,45,70],labels=["<=18", "18-25", "25-45", "45-70"])

# Inserer cette variable dans notre dataframe
data.insert(1,"intervales_ages",intervales_ages)

# Afficher le résultat
data
```

```python
data
```

```python
# Créer une table pivot qui prend en entrée le dataframe, on applique la fonction d'aggrégation count sur les maladies en indexant lse données par intervales d'age et en gardant que les colonnes sexe et activité physique
table1 = pd.pivot_table(data, columns=["sexe","alcool","fumer"], index=["maladie", "intervales_ages"], aggfunc={"maladie":"count"})
table1
```

```python

table1.to_csv("Data_diagramme_2_maladie_age_sexe_fumer_alchol.csv")
```
